```python
image_data = cv.imread('queen.jpg',1) #to read an image
```

```python
two_horizontal_queens=np.hstack((image_data,image_data)) #used to stack the sequence of input arrays
```

```python
cv2_imshow(two_horizontal_queens)
cv.imwrite('two-queens.jpg',two_horizontal_queens)
```

```python
slice_portion =two_horizontal_queens[0:rows,0:cols,0:channels]
```

```python
slice_portion.shape
```

```python
cv2_imshow(slice_portion) #to display slice portion of a image
```

```python
blurred_slice_portion = cv.blur(slice_portion,(10,10))
```

```python
cv2_imshow(blurred_slice_portion) # to get blurred image
```
